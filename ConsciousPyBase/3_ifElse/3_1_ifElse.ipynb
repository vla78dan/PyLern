Условные операторы
Условные операторы используются для принятия решений на основе различных условий. Они позволяют выполнять определенный блок кода, если условие истинно, или другой блок кода, если условие ложно. Блок кода — это набор кода, состоящий из одной или нескольких строк кода и выполняющий определённую задачу.

Условные операторы:

if ("если").
elif (сокр. "else if", "иначе если").
else ("иначе").
