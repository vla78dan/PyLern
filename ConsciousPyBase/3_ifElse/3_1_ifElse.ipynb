Условные операторы
Условные операторы используются для принятия решений на основе различных условий. Они позволяют выполнять определенный блок кода, если условие истинно, или другой блок кода, если условие ложно. Блок кода — это набор кода, состоящий из одной или нескольких строк кода и выполняющий определённую задачу.

Условные операторы:

if ("если").
elif (сокр. "else if", "иначе если").
else ("иначе").

Важные моменты:

В начале мы объявляем if и через пробел указываем условие. Условие после if может быть любым выражением, которое возвращает True или False.
В конце условия обязательно ставится двоеточие (:), это важно.
Блок кода — это любое количество кода, которое выполнится, если условие является истинным.
Блок кода начинается с отступом в четыре пробела. Отступы необходимы для обозначения логически связанного блока кода. Обычно, отступ ставится автоматически, но если забыть про двоеточие, то отступ автоматически не создастся, это для вас подсказка, не забывайте про двоеточие.
Пример 1. Операции сравнения

Как уже говорилось выше, после if вы указываете выражение, которое возвращает True или False, поэтому можно использовать выражения со сравнениями объектов:

```python
if 5 != 3:
    print("5 is not 3")

if 5 ==3:
    print("5 is 3")
```

Заключение
Условные операторы — важный элемент программирования, помогающий создавать гибкие и адаптивные программы. Важно помнить о правильном использовании двоеточий и отступов для избегания ошибок. Существуют и другие условные операторы, но давайте сначала закрепим информацию на практике, а остальные операторы разберём в следующих статьях.

Важные моменты:
Не забывайте про двоеточие. Если вы указали двоеточие и нажали на enter, то отступ в блоке кода появится автоматически. Но если забыли, то отступ не появится. Это хорошая подсказка, чтобы не забывать про двоеточие.
Отступ в 4 пробела является стандартным отступом в Python и он встречается не только в условных операторах, а также в циклах, функциях, классах, методах и т.д.
Чаще всего в условных операторах делают проверки возвращающие True или False, но бывают случаи, когда значение может быть другое. Поэтому важно знать, что любое значение отличающееся от False, 0, None и пустых объектов ("", []) является истинным в логическом контексте.
