Введение во встроенные функции Python
Функции в Python, как и методы, выполняют различные действия с объектами, например: вычисление суммы, вывод на экран, преобразование типов данных и т.д. Условно функции можно разделить на два типа:

Встроенные (Built-in Functions) — функции, разработанные и встроенные в язык Python.
Пользовательские — функции, созданные вами.
Встроенных функций в Python очень много, но для базового уровня их все изучать не обязательно. Для удобства, я оставлю ссылку на список всех встроенных функций. Далее, если я буду писать слово "функция" я буду иметь в виду встроенную функцию, пользовательские функции мы будем проходить отдельно.

```python
def sum_num(a, b):
    return a + b


result = sum_num(1, 2)
print(result)
```

Важные моменты
Условно функции можно разделить на два типа: встроенные и пользовательские.
Методы используют точку для вызова, а функции — нет.
Функции могут что-то возвращать, поэтому такие функции можно присваивать переменным или выводить на экран. Однако, есть функции, которые возвращают None и их нужно просто запускать.
Параметры и аргументы — это разные термины. Параметры — это переменные, существующие внутри функции, а аргументы — это значения, которые передаются в параметры в момент вызова функции.
Внутри функций и методов можно выполнять различные действия: указать аргументы, произвести математическую операцию, указать другую функцию или метод. Но всё сводится к тому что вы указываете какой-либо объект в качестве аргумента.
Не все функции и методы ожидают от вас аргументов. Например, если вы вызовете print() без аргументов, то на экране появится пустая строка.

abs(x)
Функция abs() используется для получения абсолютного значения числа, то есть возвращает всегда положительное число. Используется как с целыми числами, так и с вещественными. Часто используется для того, чтобы отрицательное число превратить в положительное.

```python
num = -3
print(abs(num))
```

round(number, ndigits=None)
number: Число, которое нужно округлить.
ndigits (необязательно): Количество знаков после запятой для округления. Если не указано или равно None, число округляется до ближайшего целого.
Функция round() в Python используется для округления числа до указанного количества знаков после запятой. Если параметр ndigits не указан или равен None, функция возвращает ближайшее целое число к входному значению.

```python
print(round(3.14159))
print(round(3.14159, 3))
print(round(14.14159, -1))
```

Важно добавить, что если число находится ровно посередине между двумя целыми числами, оно округляется до ближайшего чётного числа. В Python 3 функция round() использует метод округления, известный как "банковское округление" или "округление до ближайшего чётного числа":

```python
print(round(14.5))
print(round(15.5))
```

format(value, format_spec='')
Функция format() позволяет изменять внешний вид числа и возвращает строку. Функция не округляет числа, она лишь позволят красиво вывести на экран число.

Округление чисел:

".10f" – означает количество знаков, которое нужно оставить после запятой, в данном случае десять. Если чисел после запятой в оригинале меньше, то остатки заполнятся нулями (0).

```python
num = 3.14159
print(format(num, '.0f'))
print(format(num, '.1f'))
print(format(num, '.2f'))
print(format(num, '.10f'))
```

```python
num = 1000000
print(format(num, ","))  # 1,000,000
print(format(num, "_"))  # 1_000_000
```

```python
num = 42
print(format(num, "d"))  # Выведет: 42 (десятичное целое число)
print(format(num, "x"))  # Выведет: 2a (шестнадцатеричное целое число)
print(format(num, "o"))  # Выведет: 52 (восьмеричное целое число)
print(format(num, "b"))  # Выведет: 101010 (двоичное целое число)
```

divmod(a, b)
Помните мы получали результат деления нацело // и остаток от деления %, вот два этих процесса можно выполнить сразу, используя функцию divmod(). Эта функция возвращает результат деления нацело (частное) и остаток от деления,
числа a на число b в виде кортежа: (частное, остаток). Кортежи мы ещё будем проходить, если кратко, то они похожи на списки, только с круглыми скобками.

```python
result = divmod(10, 3)
print(result)  # (3, 1)

# или вот так:
частное, остаток = divmod(10, 3)
print(частное, остаток)  # 3 1
```

sorted()
sorted(iterable, /, *, key=None, reverse=False)


Функция sorted() принимает в качестве аргумента итерируемый объект (список, строку, кортеж и др.) и возвращает отсортированный список. Функция sorted() похожа на метод .sort(), но важно понимать отличия между ними:

Метод .sort() работает только со списками, а sorted() работает со строками, списками и другими итерируемыми объектами.
Метод .sort() изменяет оригинал, а функция sorted() возвращает новый объект (список). Поэтому .sort() нужно просто запустить, а sorted() можно присваивать или делать что-то ещё.

```python
lst = [3, 2, 1]
lst_sorted = sorted(lst)
print(lst)
print(lst_sorted)
```

```python
print(sorted('132'))
print(sorted('132', reverse=True))
```

```python
lst_num = [1,2,3,4,5,6,7,8,9]
result = sorted(lst_num, key=lambda x: x%2)
print(result) # мы получили список, в котором сначала идут все чётные по возрастанию, а затем нечётные по возрастанию
print(lst_num)
print(sorted(lst_num))
```

Функция range() возвращает объект, который принадлежит классу range, посмотрите на пример:

```python
print(list(range(0)))   # []
print(list(range(1)))   # [0]
print(list(range(5)))   # [0, 1, 2, 3, 4]
print(list(range(10)))
```

```python
print(list(range(0, 11, 2)))   # [0, 2, 4, 6, 8, 10] # получить, например, только чётные или нечётные числа:
print(list(range(1, 10, 2)))   # [1, 3, 5, 7, 9]
```

Функция range() сама по себе не возвращает список, поэтому мы используем функцию list(), чтобы преобразовать объект класса range в список, помните об этом.

```python
num = int(input())  # пользователь вводит число
list_num = [0, 1, 2, 3, 4, 5]
result = slice(0, num)  # используем переменную, чтобы указать конец среза
print(list_num[result])
```
